```python
u, s, vh = np.linalg.svd(raw_data)
print(s[0])
s = s**2
print(s[0])
total_value = sum(s)
weight = 0
weight_list = []
for i in range(len(s)):
    weight = weight + (s[i]/total_value)
    weight_list.append(weight)

plt.figure()
plt.plot(weight_list)
plt.xlabel('Dimension')
plt.ylabel('Cumulative explained variance percentage')

for i in range(len(weight_list)):
    if weight_list[i] > 0.7:
        print('The best intrinsic dimensionality is', i)
        print('Cumulative explained variance percentage', weight_list[i])
        break
```

上圖為Covariance matrix的eighenvalue佔其整體比例的疊加圖，而由SVD後的v^2所求出Covariance matrix的eigenvalue是由大到小排列，也就是越前面，其所佔比較最重。而原圖的部分我認為是雜訊很重的，所以在整體的前70%可以看做是最重的權重，後面則可以看做是雜訊，所以我取當所也eigenvalue累積至大於70%，當作是主要成分。其結果如下。
